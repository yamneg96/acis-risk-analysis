```python
import pandas as pd
df = pd.read_csv("MachineLearningRating_v3.txt", delimiter='|', low_memory=False)
df.info()
df.describe()

df.isnull().sum()

df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
df.groupby('Province')['LossRatio'].mean().plot(kind='bar')
```
